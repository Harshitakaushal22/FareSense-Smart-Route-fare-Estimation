```python
# Cell 1: Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look nicer
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("C:\\Users\\workk\\fare-sense-smart-route-fare-estimation\\taxi_data.csv")
df.shape
```

```python
# Cell 6: Basic statistics
df.describe()
```

```python
# Cell 7: Distribution of target variable (Trip_Price)
plt.figure(figsize=(8,4))
sns.histplot(df['Trip_Price'], bins=20, kde=True)
plt.title("Distribution of Trip Prices")
plt.show()
```

```python
# Cell 8: Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Cell 9: Price vs Distance
sns.scatterplot(x="Trip_Distance_km", y="Trip_Price", data=df)
plt.title("Price vs Distance")
plt.show()
```

```python
# Cell 10: Boxplots for categorical features vs Price
plt.figure(figsize=(8,4))
sns.boxplot(x="Traffic_Conditions", y="Trip_Price", data=df)
plt.title("Trip Price vs Traffic Conditions")
plt.show()

plt.figure(figsize=(8,4))
sns.boxplot(x="Weather", y="Trip_Price", data=df)
plt.title("Trip Price vs Weather")
plt.show()
```
